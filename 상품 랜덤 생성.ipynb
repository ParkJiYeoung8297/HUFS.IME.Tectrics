```python
#     
from korean_name_generator import namer
import random

n=100 #  


#  
code=random.sample(range(1000000000,9999999999999),n)  #10~13 

#
name=[]
for i in range(n):
    name.append(namer.generate(random.randint(0,1)))   # 0 False( ), 1 True( ))
# ,  

address=[]
detail_address=[]
for i in range(n):
    A=random.randint(0,add.shape[0]-1)
    address.append(add[" "][A])
    detail_address.append(add[""][A])
    
# 
phone=random.sample(range(11111111,99999999),100)


# ,, 
size_list={'2' :[270,100,150],
           '2-1': [350,250,100],
           '3': [340,350,210],
           '4': [410,310,280],
           '5': [480,380,340]}

counts = [17, 15, 24, 21, 23]
l,w,h=[],[],[]

keys = list(size_list.keys())

for key, count in zip(keys, counts):
    for i in range(count):
        size=size_list[key]
        l.append(size[0])
        w.append(size[1])
        h.append(size[2])
# for i in range(n):
#     size=[]
#     size=size_list[(0,len(size_list)-1)]
#     l.append(size[0])
#     w.append(size[1])
#     h.append(size[2])
remaining = n - sum(counts)
if remaining > 0:
    for i in range(remaining):
        size=size_list[keys[-1]]
        l.append(size[0])
        w.append(size[1])
        h.append(size[2])
        
indices = list(range(n))
random.shuffle(indices)

l = [l[i] for i in indices]
w = [w[i] for i in indices]
h = [h[i] for i in indices]
        
print("length:", l)
print("width:", w)
print("height:", h)
```

```python
for i in range(n):
    box=[]
    box.append(code[i])     # 
    box.append(name[i])     #
    box.append(address[i])  #
    box.append(detail_address[i])  #
    box.append("010"+str(phone[i]))    #
    box.append(l[i])        #
    box.append(w[i])        #
    box.append(h[i])        #
    df.loc[len(df)]=box     # 

df
```
